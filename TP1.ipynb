**3. Preprocessing Data**

**3.1. Data cleaning**

```python
df_train[df_train.eq("?").any(1)]
```

```python
convert = {'M':1,'B':0}
df_train["diagnosis"] = df_train["diagnosis"].map(convert)
```

Mengganti data "?" dengan null value

```python
df_train = df_train.replace(['?'],np.nan)
df_train.isna().sum()
```

```python
filler = SimpleImputer(missing_values=np.nan, strategy='mean')
df_train[:] = filler.fit_transform(df_train)
df_train
```
